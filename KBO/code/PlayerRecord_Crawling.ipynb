## 2015 - 2023 팀별 승패 기록 데이터
> 이전 기록도 가져오고 싶으면 for문 range에 연도만 수정하면 됩니다!  
(2024년은 정규시즌 진행중이어서 제외했습니다)

- '승차'가 음수인 경우는 정규시즌에서는 상위 팀보다 높은 순위였지만, 포스트시즌에서 패배하여 등수가 내려간 것입니다.

```python
def preformat_cjk(string, width, align='<', fill=' '):
    count = (width - sum(1 + (unicodedata.east_asian_width(c) in "WF") for c in string))
    return {
            '>': lambda s: fill * count + s,
            '<': lambda s: s + fill * count,
            '^': lambda s: fill * (count / 2)
                            + s
                            + fill * (count / 2 + count % 2)
    }[align](string)

kbo_data = []

# 2015년부터 2024년 팀별 기록 데이터
for year in range(2015, 2024):
    # 네이버 스포츠 기록 사이트
    baseballData = urllib.request.urlopen(f'https://sports.news.naver.com/kbaseball/record/index?category=kbo&year={year}')
    source = baseballData.read()
    baseballData.close()

    soup = BeautifulSoup(source, 'html.parser')
    soup = soup.find_all('script')

    for script in soup:
        line = str(script)

        if 'var ' in line and 'jsonTeamRecord' in line:
            recordList = line.split('jsonTeamRecord = ')

            for record in recordList:
                if record.startswith('{'):
                    recordLine = record.splitlines()
                    for finalLine in recordLine:
                        if finalLine.endswith('}]};'):
                            recordJsonData = finalLine

    r = json.loads(recordJsonData[:-1])
    regularTeamRecordList = r['regularTeamRecordList']

    order = 0
    for team in regularTeamRecordList:
        t = json.loads(str(team).replace("'", '"'))
        order += 1
        tn = preformat_cjk(t['teamName'], 10)
        won = t['won']
        lost = t['lost']
        winDiff = t['winDiff']
        recentResult = t['recentResult']

        kbo_data.append({
            '연도': year, '순위': order, '팀 이름': tn, '승': won, '패': lost, '승차': winDiff, '최근 경기 결과': recentResult
        })

kbo_results = pd.DataFrame(kbo_data)
kbo_results
```

```python
# 2015-2024 팀별 승패 기록 데이터 csv 저장
kbo_results.to_csv('kbo_results_2015_to_2023.csv', index=False, encoding="utf-8-sig")
```

> 한국 야구 팀만 선별하여 인덱스로 만들고 컬럼을 날짜로 지정하여 총 검색량을 데이터로 삽입하는 과정

```python
# 데이터가 저장된 디렉토리 경로
data_dir = '/content/drive/MyDrive/Colab Notebooks/[DArt-B] 야구 데이터분석 프로젝트/data/baseball_trend'

# KBO 리그만 필터링
teams = ['두산베어스', 'LG트윈스', '한화이글스', 'SSG랜더스', '롯데자이언츠', '삼성라이온즈',
         '기아타이거즈', '키움히어로즈', 'NC다이노스', 'KT위즈']

# 날짜 범위
date_range = pd.date_range(start='2024-03-23', end='2024-05-24', freq='D')
trend_2024 = pd.DataFrame(index=teams, columns=date_range)

for single_date in date_range:
    file_name = f'DM_BSBL_CLUB_TOP10_TREND_DATA_RESULT_{single_date.strftime("%Y%m%d")}.csv'
    file_path = os.path.join(data_dir, file_name)

    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        df_filtered = df[df['SRCHWRD_NM'].isin(teams)][['SRCHWRD_NM', 'SCCNT_SM_VALUE']]

        for index, row in df_filtered.iterrows():
            trend_2024.loc[row['SRCHWRD_NM'], single_date] = row['SCCNT_SM_VALUE']

trend_2024
```

```python
# 2024년 트렌드 데이터 csv 저장
trend_2024.to_csv('trend_2024.csv', index=False, encoding="utf-8-sig")
```
